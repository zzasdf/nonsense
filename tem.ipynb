```python
agg_ops = ['', 'MAX', 'MIN', 'COUNT', 'SUM', 'AVG']
cond_ops = ['=', '>', '<', 'OP']
syms = ['SELECT', 'WHERE', 'AND', 'COL', 'TABLE', 'CAPTION', 'PAGE', 'SECTION', 'OP', 'COND', 'QUESTION', 'AGG', 'AGGOPS', 'CONDOPS']
```

```python
def to_sql(question, table):
    def get_conds(conds):
        if len(conds)==0:
            return ""
        re="WHERE "
        for i, member in enumerate(conds):
            if i!=0:
                re+='AND '
            re+="col{}".format(member[0])+" "+cond_ops[member[1]]+" col{}".format(member[0])+' '
        return re
    def sub_to_sql(q):
        re="SELECT "
        tid=q["table_id"]
        sel=q["sql"]["sel"]
        conds=q["sql"]["conds"]
        if q["sql"]["agg"]!=0:
            re+=agg_ops[q["sql"]["agg"]]+'('
            re+="col{}".format(sel)
            re+=')'
        else:
            re+="col{}".format(sel)
        re+=" FROM table_{} ".format(tid.replace('-',"_"))
        re+=get_conds(conds)
        return re
    re=[sub_to_sql(member) for member in question]
    return re
        
```

```python
lis=to_sql(question, table)
```
